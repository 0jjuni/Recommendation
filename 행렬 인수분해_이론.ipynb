## 값을 모르게
- 누락값이 있는 부분을 맞추는 것

```python
A = np.array([[4, 1, 5, np.nan, 1],
              [2, 3, np.nan, 2, 3],
              [1, np.nan, 4, 1, 3],
              [np.nan, 2, 4, np.nan, 2],
              [1, np.nan, 4, 1, 3]])
A
```

```python
mask = ~np.isnan(A)
mask
```

```python
# U = 5 * ?
# V = ? * 5
# ?가 커질 수록 파라미터 수가 많아진다.
```

```python
k = 2
m, n = A.shape

# 아무 값을 줘놓고 경사하강법을 통해 수렴해 나갈 것임.
U = np.random.rand(m, k)
V = np.random.rand(n, k)  # k행 n열이 맞음. 계산을 위해 이렇게 표현

U @ V.T
```

```python
lr = 0.01  # 학습률

for i in range(5000):
    A_pred = U @ V.T

    # 오차마스크
    E = np.zeros((m, n))
    E[mask] = A_pred[mask] - A[mask]  # True인 부분만 접근한다.

    U_grad = E @ V
    V_grad = E.T @ U

    U -= U_grad * lr
    V -= V_grad * lr

    loss = sum(E[mask] **2)
    print(loss)
```

```python
A_pred
```

```python
# nan값을 경사하강법을 통한 예측값을 통해 채우기
A_filled = A.copy()

A_filled[~mask] = A_pred[~mask]
A_filled
```

```python
# 컴퓨터가 예측해준 값들만 넣어 둠.
zeros = np.zeros(A_pred.shape)
zeros[~mask] += A_pred[~mask]
zeros
```

```python
for i in range(len(zeros)):
    a = zeros[i]
    for j in range(len(a)):
        if a[j] >= 4:
            print(f"{i}번째 유저, {j}번째 영화 추천!")
```
